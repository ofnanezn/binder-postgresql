### Relación entre Price y Year

El objetivo de analizar dichas variables es el de identificar si existe algún patrón en el comportamiento del precio de los vehículos a medida en que se reduce el tiempo de haber sido fabricado. Lo esperado es que exista una relación positiva, es decir, que el precio aumente, en promedio, para los vehículos que son más recientes. Asimismo, identificar si a nivel univariado existe algún tipo de problema con la distribución.

* ¿Nota algo extraño en los rangos de las variables usadas? Si es así, proponga alguna forma para corregir dicho problema.

El rango de la variable **year** es inconsistente, ya que tiene valores de 0, lo cual hace que las gráficas se vean concentradas hacia la derecha. Una forma de solucionar el problema es validar la cantidad de registros que tienen esos valores y si no impacta de forma significativa el tamaño del dataset se podría eliminar. Los siguientes resultados muestran que hay valores menores a 1950, los cuales no son representativos, en cantidad de registros. Los resultados de haber truncado los valores de la variable se ven a continuación. Adicionalmente, se encuentra un problema con el precio de los vehículos registrados, ya que hay para todos los años hay precios de 0, lo cual no tiene sentido. Haciendo un análisis de la distribución de dicha variable se encuentra que cerca del 10% de los datos totales se encuentran en 0, por lo cual se filtran dichos valores. Asimismo, hay valores que son muy altos y los cuales están afectando la distribución, por lo cual se eliminan también del dataset.

```python
data = data[data["year"] >= 1990]
data = data[data["price"] >= np.percentile(data["price"], 10)]
data = data[data["price"] <= np.percentile(data["price"], 99.9)]
data.columns = [item.capitalize() for item in data.columns.tolist()]
data["Manufacturer"] = data["Manufacturer"].astype(str)
data["Manufacturer"] = [item.capitalize() for item in data["Manufacturer"]]

print("""Luego de las transformaciones hechas para corregir algunos de los problemas con las variables price y year el tamaño del dataset es de {}, \
es decir, {} registros menos a los obtenidos luego del cruce.""".format(data.shape, 536792-data.shape[0]))
```

```python
plt.style.use("dark_background")
fig, ax = plt.subplots(facecolor="#14161B", 
                       figsize=(28, 12))

ax.set_xlabel('Price')
ax.set_ylabel('Manufacturer')
ax.set_title("")
ax.set_facecolor("#14161B")

data[["Price", "Year"]].plot(kind="scatter", 
                             x="Year", 
                             y="Price", 
                             figsize=(28, 10), 
                             ax=ax,
                             color="#58A4B0",
                             title="Dispersión de precios por año.", 
                             fontsize=12, 
                             grid=False)
```

Se encontró que existe una relación creciente entre el precio y el añó de fabricación de los vehículos, como era de esperarse, es decir, los precios tienden a ser más altos cuando ha transcurrido menos tiempo desde su fabricación. Asímismo, hay una alta dispersión en vehículos antiguos, los cuales alcanzan valores cercanos a los más recientes, lo cual puede implicar la venta de vehículos clásicos y de alto valor.

### Relación entre Price y Manufacturer

El objetivo de analizar estas dos variables es la identificar patrones asociados con el fabricante y el precio registrado del vehículo. En este caso se espera que para las marcas premium haya unos rangos de precios más altos que para las marcas de gama baja o media.

```python
plt.style.use("dark_background")
fig, ax = plt.subplots(facecolor="#14161B", 
                       figsize=(28, 12))

ax.set_xlabel('Price')
ax.set_ylabel('Manufacturer')
ax.set_title("")
ax.set_facecolor("#14161B")
props = dict(boxes="#FE3509", whiskers="#58A4B0", medians="#FE3509", caps="#FE3509")

data[["Price", "Manufacturer"]].sort_values(by=["Manufacturer", "Price"], ascending=False).boxplot(column=["Price"], 
                                                                                  by="Manufacturer",
                                                                                  vert=False, 
                                                                                  ax=ax, 
                                                                                  color=props, 
                                                                                  showfliers=False, 
                                                                                  showmeans=True, 
                                                                                  fontsize=12, 
                                                                                  grid=False)
```

Mediante la gŕafica anterior se identificó que hay una diferencia entre los precios de vehículos de algunas marcas en específico, lo cual es un indicio de vehículos de segmentos premium o, unido a la conclusión anterior, vehículos clásicos de marcas reconocidas y que tienen valores altos. ### Correlación entre variables númericas

```python
corr = data.select_dtypes(include=np.number)[["Price", "Year", "Number of crashes", "Poverty"]].corr()

plt.style.use("dark_background")
fig, ax = plt.subplots(facecolor="#14161B", 
                       figsize=(20, 10))

ax.set_xlabel('Price')
ax.set_ylabel('Manufacturer')
ax.set_title("Correlaciones entre variables numéricas")

sns.set(font_scale=1.2)
sns.heatmap(corr,
            linewidths=0.5, 
            linecolor="#14161B",
            vmin=-1, 
            vmax=1, 
            center=0,
            cmap="YlGnBu",
            square=True, 
            ax=ax, 
            annot=True, 
            annot_kws={"color": "white"},
            cbar_kws={"orientation": "vertical"})

cax = plt.gcf().axes[-1]
cax.tick_params(color="white", labelcolor="white")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline, FeatureUnion

y = data["Price"]
X = data.drop("Price", axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=.7, test_size=0.3, random_state=1234)

to_encode = ["Fuel", "Year", "Title_status", "Transmission", "State"]
[data[item].replace(np.nan, None, inplace=True) for item in to_encode]

Fuel_Encoder = OneHotEncoder(handle_unknown="error", sparse=False, drop="first")
Year_Encoder = OneHotEncoder(handle_unknown="error", sparse=False, drop="first")
Title_Encoder = OneHotEncoder(handle_unknown="error", sparse=False, drop="first")
Transmission_Encoder = OneHotEncoder(handle_unknown="error", sparse=False, drop="first")
State_Encoder = OneHotEncoder(handle_unknown="error", sparse=False, drop="first")

encoders = [Fuel_Encoder, Year_Encoder, Title_Encoder, Transmission_Encoder, State_Encoder]

for var, encoder in zip(to_encode, encoders):
    encoder.fit(np.array(X_train[var]).reshape(-1, 1))
    new_columns = [str(item) + "_Encoded" for item in encoder.categories_[0].tolist()[1:]]
    X_train = pd.concat([X_train, 
                         pd.DataFrame(encoder.transform(np.array(X_train[var]).reshape(-1, 1)), 
                                      columns=new_columns, 
                                      index=X_train.index)],
                        axis=1)
    X_test = pd.concat([X_test, 
                        pd.DataFrame(encoder.transform(np.array(X_test[var]).reshape(-1, 1)), 
                                     columns=new_columns, 
                                     index=X_test.index)],
                       axis=1)

X_test.head()
```

```python
X_train_Encoded = X_train.loc[:, [item.endswith("_Encoded") for item in X_train.columns]]
X_test_Encoded = X_test.loc[:, [item.endswith("_Encoded") for item in X_test.columns]]

MyLinear_Regression = LinearRegression()

Linear_Regression = Pipeline([("MyLinear_Regression", MyLinear_Regression)])
Linear_Regression.fit(X_train_Encoded, y_train)
X_train_Score = Linear_Regression.score(X_train_Encoded, y_train)
X_test_Score = Linear_Regression.score(X_test_Encoded, y_test)
print("Training Score: {} and Testing Score: {} ")
```

```python
from sklearn.ensemble import RandomForestRegressor

Random_Forest = RandomForestRegressor()
Random_Forest.fit(X_train_Encoded, y_train)
X_train_Score = Random_Forest.score(X_train_Encoded, y_train)
X_test_Score = Random_Forest.score(X_test_Encoded, y_test)
print("Training Score: {} and Testing Score: {} ")
```
